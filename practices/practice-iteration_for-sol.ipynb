**문제 2**

다음 모양을 출력하는 코드를 작성하라.
두 개의 별 기호 사이에 공백(space)가 사용됨에 주의한다.
공백도 하나의 기호로 취급된다.
단, `print()` 함수와 `for` 반복문은 한 번씩만 활용한다.

```
        * 
      * * 
    * * * 
  * * * * 
* * * * * 
  * * * * 
    * * * 
      * * 
        * 
```

힌트: `range()` 함수 활용

답:

`for` 반복문을 `num` 변수가 1부터 9까지에 대해 실행하도록 한다.
이를 위해 `range(1, 10)` 함수 표현식을 이용한다.

````python
for num in range(1, 10):
    if num <= 5:
        spaces = 5 - num
        stars = num
    else:
        spaces = num - 5
        stars = 10 - num
        
    print("  " * spaces + "* " * stars)
````

**문제 5**

다음과 같이 영어 알파벳으로 구성된 문자열이 주어졌다.

````python
text = "python data analysis"
````

(1) 문자열에 포함된 모음의 개수를 확인하는 코드를 작성하려 한다.
단, 모음 알파벳은 아래 문자열에 포함되어 있다.

````python
vowels='AEIOUaeiou'
````

예를 들어 다음과 같이 `for` 반복문을 이용할 수 있다.

````python
count_vowels = 0
for char in text:
    if char in vowels:
        count_vowels += 1

print("모음 개수:", count_vowels)
````

`continue` 명령문을 사용하여 위 코드를 수정하라. 
단 동일한 결과가 나와야 한다.

답:

모음이 아닐 때 반복문의 다음 경우를 처리하도록 하면 된다.

````python
count_vowels = 0
for char in text:
    if char not in vowels:
        continue
    
    count_vowels += 1

print(count_vowels)
````

(2) 문자열에 포함된 자음의 개수를 확인하는 코드를 작성하려 한다.

`text` 문자열은 공백을 포함하는데, 공백은 알파벳이 아니다.
따라서 모음과는 달리 자음만 추출하려면 공백도 제외시켜야 한다.
이를 위해 문자열이 영어 26개의 알파벳으로만 구성되었는지 여부를 판정하는 `isalpha()`
문자열 메서드를 이용한다.

````python
'a'.isalpha()
````

````python
'ab'.isalpha()
````

````python
'ab3'.isalpha()
````

아래 코드는 주어진 영어 문장 `text`에 대해 모음이 아닌 알파벳이 사용된 횟수를
`for` 반복문을 이용하여 계산한다.
`isalpha()` 메서드를 이용하여 모음이 아니면서 영어 알파벳인 경우일 때만 `count_consonant` 값이 업데이트됨에 주의한다.

````python
count_consonant = 0
for char in text:
    if (char not in vowels) and char.isalpha():
        count_consonant += 1

print("자음 개수:", count_consonant)
````

`continue` 명령문을 사용하도록 위 코드를 수정하라.

답:

기호가 모음 알파벳이거나 알파벳이 아니면 무시하도록
`continue` 명령문을 활용한다.

````python
count_consonant = 0
for char in text:
    if (char in vowels) or not char.isalpha():
        continue
        
    count_consonant += 1

print("자음 개수:", count_consonant)
````

`text`가 가리키는 문자열은 모듬 6개, 자음 12개, 그리고 공백이 2개, 총 20개의 문자로 구성되었으며
아래 코드가 이를 확인해준다.

````python
len(text) == count_vowels + count_consonant + 2
````

**문제 6**

두 개의 문자열이 다음과 같이 주어졌다.

````python
string = "PythonForPythonForPythonForPythonForPython"
substring = "ForPythonFor"
````

이때 `substring`이 `string`에 사용된 횟수는 두 번으로 확인된다.

````python
string.count(substring)
````

이유는 `count()` 메서드는 지정된 문자열이 겹쳐지지 않는 경우만 확인하기 때문이다.

(1) `substring`이 겹쳐지는 것도 허용하면서 사용된 횟수를 확인하면 3번이다.
이 사실을 확인하는 코드를 `for` 반복문을 이용하여 작성하라.

힌트: 문자열 슬라이싱 활용

답:

먼저 아래 두 변수를 선언한다.

- `count`: `substring` 문자열이 사용되는 횟수 카운팅
- `pos`: `string` 문자열을 대상으로 `for` 반복문을 실행할 때 문자열의 인덱스 역할 수행

```python
count = 0
pos = 0
```

이제 `char` 변수가 `string` 문자열에 포함된 각각의 문자를 가리키는 변수로 활용한다.
또한 `char` 변수가 가리키는 문자가 있는 곳에서부터 `substring` 문자열의 길이만큼 슬라이싱을 적용한 결과가
`substring`과 일치하는지 여부에 따라 `count` 값을 1만큼 증가시킨다.
이때 `pos`가 `char`의 인덱스 역할을 수행하도록 하여 슬라이싱을 적용하기에
`for` 반복문이 한 번 실행될 때마다 1씩 증가시킨다.

````python
count = 0 # 카운팅
pos = 0   # 인덱스 역할

for char in string:
    sub = string[pos: pos+len(substring)] # 슬라이싱

    if substring == sub: # 슬라이싱으로 확인된 문자열이 맞는 경우
        count += 1
    
    pos += 1
    
print(f"{substring}이 중첩되어 사용됫 횟수:", f"{count}회")
````

(2) 위 코드를 아래 조건이 만족되도록 수정하라.

- `count` 변수가 2가 될 때까지만 `for` 반복문이 실행되어야 한다.
- 다음과 같이 출력되어야 함.

        ForPythonFor이 중첩되어 사용됫 횟수: 최소 2회

힌트: `break` 명령문 활용

답:

````python
count = 0 # 카운팅
pos = 0   # 인덱스 역할

for char in string:
    sub = string[pos: pos+len(substring)] # 슬라이싱

    if substring == sub: # 슬라이싱으로 확인된 문자열이 맞는 경우
        count += 1
        
    if count == 2:
        break
    
    pos += 1
    
print(f"{substring}이 중첩되어 사용됫 횟수:", f"최소 {count}회")
````

**문제 7**

````python
for n in range(2, 10):
    is_prime = True # 이 값이 계속 유지되면 소수임

    for x in range(2, n):
        if n % x == 0:
            is_prime = False
            print("-", n, "=", x, "*", n//x)
            break
            print("따라서", n, "은(는) 소수 아님!")

    continue
    
    if is_prime:
        print("-", n, "은(는) 소수다.")
    else:
        print(" ", n, "은(는) 소수가 아니다.")
````

        4번 줄에서 시작하는 내부의 `for` 반복문 실행 종료
    - 곧바로 `continue` 명령문이 실행되어 1번 줄의 `for` 반복문을 `n+1`에 대해 실행.

- 변수 `n`이 소수를 가리킬 때
    - 5번 줄에서 시작하는 `if ...` 조건문의 본문이 실행되지 않아 `break` 명령문이 전혀 실행되지 않음.
    - 따라서 `for` 반복문이 정상적으로 실행 종료됨.

    1번 줄의 `for` 반복문을 `n+1`에 대해 실행을 이어감.
- 따라서 13번 줄부터 시작하는 `if ... else ...` 명령문은 어떤 경우에도 실행되지 않게 되어

- 바깥쪽 `for` 반복문에서; 입력값의 배수가 아닌 단은 건너뛰도록 할 때.

````python
one2nine = range(1, 10)

table_index = int(input("단수를 입력하세요: "))

print(f"\n{table_index}의 배수 구구단 출력\n")
for i in one2nine:
    if i % table_index != 0:
        continue
        
    print(f"{i}단\n")
    for j in one2nine:
        if j % table_index == 0:
            continue
        print(f"{i} * {j} = {i * j:>2d}")
        
    print()
````

몫과 나머지를 연속적으로 적용해서 확인할 수 있다.
이를 `for` 반복문으로 구현하면 다음과 같다.

````python
n = 1730

coin_number = 0 # 사용된 동전 개수

for coin in [500, 250, 100, 50, 10]:
    count = n // coin
    coin_number += count
    n = n % coin

print(f"1,730원 지불에 필요한 동전 수: {coin_number}개")
````

620원을 지불하려면 동전 4개가 필요하다.

````python
n = 620

coin_number = 0 # 사용된 동전 개수

for coin in [500, 250, 100, 50, 10]:
    count = n // coin
    coin_number += count
    n = n % coin

print(f"620원 지불에 필요한 동전 수: {coin_number}개")
````

`print()` 함수를 적절히 이용해서 거스름돈 지불에 사용된 동전을 구체적으로 확인할 수 있다.

````python
n = 1730
print(f"{n}원 지불에 필요한 동전 수\n")

coin_number = 0 # 사용된 동전 개수

for coin in [500, 250, 100, 50, 10]:
    count = n // coin
    print(coin, "원 동전:", count, "개")            
    coin_number += count
    n = n % coin
````
