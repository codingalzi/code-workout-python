**문제 1**

다음 역삼각형 모양을 출력하는 코드를 작성하라.
두 개의 별 기호 사이에 공백(space)가 사용됨에 주의한다.
공백도 하나의 기호로 취급된다.
단, `while` 반복문을 이용한다.

```
* * * * *  
* * * *  
* * *  
* *  
*
```

답:

아래 코드는 `num` 변수를 5부터해서 1씩 줄이며 필요한 만큼의 별표 기호를 출력한다.

````python
num = 5

while num >= 0:
    print("* " * num)
    num -= 1
````

**문제 2**

사용자로부터 두 개의 양의 정수를 입력 받아
양의 정수를 다른 양의 정수로 나눴을 때의 
몫과 나머지를 계산하는 코드를 `while` 반복문을 이용하여 구현하라.
또한 아래 예시처럼 계산 결과를 화면에 출력되도록 한다.

```python
45 을(를) 7 (으)로 나눴을 때의 몫: 6
45 을(를) 7 (으)로 나눴을 때의 나머지: 3
```
    
단, 연산자는 덧셈, 뺄셈, 곱셈 연산자만 사용해야 한다. 
즉, 몫 연산자, 나머지 연산자 등 파이썬에서 기본으로 제공하는 
연산자나 함수는 사용하지 않는다.

답:

예를 들어 45을 7로 나눌 때의 나머지는 7단을 외우다가 45보다 커지는 순간에
그때까지 7에 곱한 값에서 1을 뺀 값이 몫이고,
몫과 7을 곱한 값과 45와의 차이가 나머지가 된다.

```
7 * 1 = 7
7 * 2 = 14
...
7 * 6 = 42
7 * 7 = 49   <= 45보다 큼
```

즉, 7에서 1을 뺀 6이 몫이고 6을 7과 곱한 값인 42와 45의 차이인 3이 나머지가 된다.

이 아이디어를 코드로 구현하면 다음과 같다.

- `number`: 나눠지는 수를 인자로 받는 매개변수
- `divisor`: 나누는 수를 인자로 받는 매개변수
- `count`: 나누는 수에 곱해지는 수
- `quotient`: `number // divisor`
- `remainder`: `number % divisor`

그러면 `divisor * count <= number`이 만족되는 동안 `count` 값을 1씩 키운다.

````python
number, divisor = input("두 개의 양의 정수를 공백으로 구분하여 입력: ").split()
number = int(number)
divisor = int(divisor)

count = 1
while divisor * count <= number:
    count += 1

quotient = count - 1
remainder = number - divisor * quotient

print(f"{number}을(를) {divisor}(으)로 나눴을 때의 몫: {quotient}")
print(f"{number}을(를) {divisor}(으)로 나눴을 때의 나머지: {remainder}")
````

**문제 3**

복리 정기예금과 관련해서 원금, 연 복리이자율, 예치기간이 주어졌을 때 
지정된 예치기간 후에 받을 원리합계, 즉, 원금과 이자의 합(원리합계)을 계산하는 코드를 구현하라.
단, 총이자를 계산할 때 `while` 반복문을 반드시 이용한다.

예를 들어, 연 복리이자율 5%인 정기예금에 백만원을 3년간 은행에 맡겼을 때
3년 후에 수령할 금액은 원금 백만원에 이자 157,625원이 추가되어
총 1,157,625원이 되며,
화면에 다음과 같이 출력되어야 한다.

```
1000000원을 복리이자 5.0%로 3년 예치 후 받는 원금 + 이자는 1157625.0원입니다.
```

힌트: 복리이자의 원리합계 계산 코드는 다음과 같다.

```
total = principal * (1 + rate)**period
```

- `total`: 원리합계
- `principal`: 원금
- `rate`: 복리이자율
- `period`: 년 단위의 기간

단, 원금, 복리이자율, 기간은 다음과 같다.

````python
principal = 1_000_000
rate = 0.05
period = 3
````

답:

복리에 대한 원리합계를 계산하는 식은 다음과 같았다.

```python
total = principal * (1 + rate)**period
```

따라서 저축기간인 `period` 가 1년 증가할 때마다 원리합계는 이전 원리합계에 `(1 + rate)`를 곱한 결과이다.
이 정보를 `while` 반복문에 적용하여 다음과 같이 원리합계를 계산할 수 있다.

````python
total = principal # period가 0일 때는 원금만 있음.
num_of_years = 1

while num_of_years <= period:
    total = total * (1 + rate)
    num_of_years = num_of_years + 1

print(f"{principal}원을 복리이자 {rate*100}%로 {period}년 예치 후 받는 원금 + 이자는 {total}원입니다.")
````

**문제 4**

아래 코드의 실행결과를 설명하라.

````python
n = 3
while n < 10:
    is_even = True

    if n % 2 == 1:
        is_even = False
        print("-", n, "은(는) 홀수다.")
    
    break

    if is_even:
        print("-", n, "은(는) 짝수다.")
        
    n += 1
    
print("실행 종료!")
````

답:

`while` 반복문은 `n`이 3부터 10까지의 정수를 차례대로 가리키는 동안 실행된다.

- 5번 줄에서 시작하는 `if` 조건문은 `n`이 홀수일 때 홀수임을 확인한다.
- 이후에 9번 줄에서 `break` 명령문이 실행되어 `while` 반복문의 실행이 중단된다.
- 결국 `n`이 3인 경우에 홀수임을 확인한 후에 곧바로 `while` 반복문의 실행이 중단된다.
- `while` 반복문이 종료된 후에 이어지는 명령문이 계속해서 실행되어
    마지막 줄의 `print("실행 종료!")`에 의해 코드의 실행이 종료되었음을 확인해준다.

**문제 5**

사용자로부터 1부터 9사이의 정수를 입력 받은 후 입력된 정수의 배수에 해당하는 구구단만 
지정된 형식으로 화면에 출력하는 코드를 작성하라.

예를 들어, 사용자가 3을 입력하면 3의 배수인 3, 6, 9에 대한 구구단만 아래와 같이 출력해야 한다.
또한 `while` 반복문과 함께 `continue`, `break` 명령문을 구구단 출력에 이용해야 하며,
곱셈 결과는 오른쪽으로 정렬되어야 한다.
단, `for` 반복문은 절대 사용할 수 없다.

- 3을 입력하는 경우

```
단수를 입력하세요: 3

3의 배수 구구단 출력

1단

1 * 3 =  3
1 * 6 =  6
1 * 9 =  9

2단

2 * 3 =  6
2 * 6 = 12
2 * 9 = 18
```

- 4를 입력하는 경우

```
단수를 입력하세요: 4

4의 배수 구구단 출력

1단

1 * 4 =  4
1 * 8 =  8

2단

2 * 4 =  8
2 * 8 = 16

3단

3 * 4 = 12
3 * 8 = 24
```

답:

다음 두 경우에 `continue` 명령문과 `break` 명령문이 실행되도록 해야 한다.

- 바깥쪽 `while` 반복문에서; 입력값보다 큰 단은 실행되지 않도록 할 때 `break` 명령문 실행
- 안쪽 `while` 반복문에서: 입력값의 배수가 아닌 수에 대한 곱셈은 무시되도록 할 때 `continue` 명령문 실행

- 3을 입력하는 경우

````python
one2nine = range(1, 10)

table_index = int(input("단수를 입력하세요: "))

print(f"\n{table_index}의 배수 구구단 출력\n")

i = 1
while i in one2nine:
    if i % table_index == 0:
        break
        
    print(f"{i}단\n")
    j = 1
    while j in one2nine:
        if j % table_index != 0:
            j += 1
            continue
        print(f"{i} * {j} = {i * j:>2d}")
        j += 1
        
    i += 1
        
    print()
````

- 4를 입력하는 경우

````python
one2nine = range(1, 10)

table_index = int(input("단수를 입력하세요: "))

print(f"\n{table_index}의 배수 구구단 출력\n")

i = 1
while i in one2nine:
    if i % table_index == 0:
        break
        
    print(f"{i}단\n")
    j = 1
    while j in one2nine:
        if j % table_index != 0:
            j += 1
            continue
        print(f"{i} * {j} = {i * j:>2d}")
        j += 1
        
    i += 1
        
    print()
````

**문제 6**

아래 두 코드의 실행 결과가 동일한 이유를 설명하라.

- 코드 1

````python
for i in range(4):
    j = 0
    while j < 4:
        if i < j:
            break
        print([i, j])
        j += 1
````

- 코드 2

````python
i = 0
while i in range(4):
    for j in range(4):
        if i < j:
            continue
        print([i, j])
        
    i += 1
````

답:

코드 1은 다음과 같이 작동한다.

- `for` 반복문에 의해 `i`가 0부터 3까지 변함.
- 각각의 `i`에 대해 `i < j` 가 거짓인 경우, 즉 `j <= i`가 참인 경우에만 
    리스트 `[i, j]` 출력하고 `j`를 1만큼 증가시키고 `while` 명령문 반복.
- 각각의 `i`에 대해 `i < j` 가 참이 되는 순간 `break` 명령문이 실행되어
    `while` 반복문이 종료되어, `for` 반복문에서 `i`의 다음 경우로 넘어감.
- 결론적으로 `i >= j`인 경우에만 `[i, j]` 출력

코드 2는 다음과 같이 작동한다.

- `while` 반복문에 의해 `i`가 0부터 3까지 변함.
- 각각의 `i`에 대해 `i < j` 가 거짓인 경우에만 리스트 `[i, j]` 출력하고 `for` 반복문에서 `j`의 다음 경우로 넘어감..
- 각각의 `i`에 대해 `i < j` 가 참이면 `continue` 명령문이 실행되어 `for` 반복문에서 `j`의 다음 경우로 넘어감..
- 결론적으로 `i >= j`인 경우에만 `[i, j]` 출력.
