단, 총이자를 계산할 때 `while` 반복문을 반드시 이용한다.

예를 들어, 연 복리이자율 5%인 정기예금에 백만원을 3년간 은행에 맡겼을 때
3년 후에 수령할 금액은 원금 백만원에 이자 157,625원이 추가되어

- `period`: 년 단위의 기간

단, 원금, 복리이자율, 기간은 다음과 같다.

**문제 4**

아래 코드의 실행결과를 설명하라.

````python
n = 3
while n < 10:
    is_even = True

    if n % 2 == 1:
        is_even = False
        print("-", n, "은(는) 홀수다.")
    
    break

    if is_even:
        print("-", n, "은(는) 짝수다.")
        
    n += 1
    
print("실행 종료!")
````

**문제 6**

아래 두 코드의 실행 결과가 동일한 이유를 설명하라.

- 코드 1

````python
for i in range(4):
    j = 0
    while j < 4:
        if i < j:
            break
        print([i, j])
        j += 1
````

- 코드 2

````python
i = 0
while i in range(4):
    for j in range(4):
        if i < j:
            continue
        print([i, j])
        
    i += 1
````
