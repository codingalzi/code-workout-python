# (연습) 상속

**주의 사항**

* 기존에 작성된 코드 셀과 텍스트 셀은 수정하지 않는다.
* 필요한 경우 코드 셀 또는 텍스트 셀을 추가해서 사용한다.
* 실습 파일을 제출할 때 구글 드라이브의 링크를 이용하여 누구나 공유할 수 있도록 설정한다.

**문제 1**

`Character` 클래스에서 사용되는 모든 인스턴스 속성에 대한 게터/세터 메서드를 선언한 후에 
`attack()` 메서드를 게터/세터 메서드만을 사용하여 선언하라. 
즉, 인스턴스 속성 변수를 직접 사용하지 않아야 한다.

**문제 2**

구성 방식을 이용하여 `Character` 클래스가 보다 다양한 기능을 제공하도록 만들어라.
예를 들어, 캐릭터가 사용하는 무기를 동적으로 변경할 수 있도록 만들 수 있다.
