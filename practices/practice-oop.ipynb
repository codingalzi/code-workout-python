**문제 1**

`polygon(t, length, n)` 함수는 한 변의 길이가 `length`인 정 n-각형을 그린다.
단, `t`는 거북이 객체를 가리킨다.

```python
def polygon(t, length, n):
    turn = 360/n  # 회전 각도
    for _ in range(n):
        t.forward(length)
        t.left(turn)
```

`polygon()` 함수를 이용하여 원을 그리는 `circle(t, r, n)` 함수를 구현하여라.
세 매개 변수 각각에 입력되는 값의 의미는 다음과 같다.

- `t`: 거북이 객체
- `r`: 원의 반지름
- `n`: 정 n-각형을 활용한 원 그리기.

예를 들어 `circle(t, 100, n=100)`를 실행하면 아래 모양의 원이 그려져야 한다.
