```python
re.search('[cmf].{2}', 'dan')
```

```python
re.search(r'[cmf]\w+', 'ran')
```

```python
re.search('[cmf].{2}', 'pan')
```

- 특정 문자는 제외시켜야 하는 경우: d, r, p 로만 시작하면 안됨.

```python
re.search('[^drp]an', 'dan')
```

```python
re.search('[^drp]an', 'ran')
```

```python
re.search('[^drp]an', 'pan')
```

```python
re.search('[^drp]an', 'can')
```

**문제 4**

아래 표에 포함된 각각의 문자열에 대해 성사로 표시된 문자열 전체를 매칭 시키는 하나의 정규표현식을
선언한 다음에 이를 확인하라.
단, 미성사로 표시된 문자열에 대해서는 매칭이 없어야 한다.

힌트: `\s`는 모든 화이트 스페이스 문자를 가리킨다.
반면에 `\S`는 화이트 스페이스를 제외한 임의의 문자를 가리킨다.

| 매칭 | 문자열 |
| :---: | :--- |
| 성사 | `1. abc` |
| 성사 | `2.  abc` |
| 성사 | `3.   abc` |
| 미성사 | `4.abc` |

```python
re.search(r'\d\.\s+abc', '1. abc')
```

```python
re.search(r'\d\.\s+abc', '2.  abc')
```

```python
re.search(r'\d\.\s+abc', '3.   abc')
```

```python
re.search(r'\d\.\s+abc', '4.abc')
```

```python
re.search(r'\d\.\S', '4.abc')
```

**문제 5**

아래 표에 포함된 각각의 문자열에 대해 성사로 표시된 문자열 전체를 매칭 시키는 하나의 정규표현식을
선언한 다음에 이를 확인하라.
단, 미성사로 표시된 문자열에 대해서는 매칭이 없어야 한다.

힌트: 부분문자열을 캡처한 뒤에 재활용할 수 있는 기능도 지원한다.
캡처하고자 하는 부분문자열의 패턴을 소괄호로 감싸인 정규표현식으로 지정한다.

| 그룹 지정 | 문자열 | 그룹 |
| :---: | :--- | :--- |
| 성사 | `file_record_transcript.pdf` | `file_record_transcript` |
| 성사 | `file_07241999.pdf` | `file_07241999` |
| 미성사 | `testfile_fake.pdf.tmp` | |

답:

```python
match = re.search(r'^(file.+)\.pdf$', 'file_record_trasncript.pdf')
match
```

- `group(0)` 반환값: 지정된 정규표현식에 매칭된 부분문자열 전체

```python
match.group(0)
```

- `group(1)` 반환값: 캡처되어 첫째 그룹으로 지정된 부분문자열

```python
match.group(1)
```

여러 개의 캡처를 지정할 수도 있다.
아래 코드는 파일명뿐만 아니라 확장자도 캡처한다.

```python
match = re.search(r'^(file.+)\.(pdf)$', 'file_record_trasncript.pdf')
match
```

`re.Match` 객체의 `group()` 메서드의 0과 1에 대한 반환값은 이전과 동일하다.

```python
match.group(0)
```

```python
match.group(1)
```

추가로 둘째 그룹으로 캡처된 부분문자열도 함께 저장된다.

```python
match.group(2)
```

**문제 6**

아래 표에 포함된 각각의 문자열에 대해 성사로 표시된 문자열 전체를 매칭 시키는 하나의 정규표현식을
선언한 다음에 이를 확인하라.
단, 미성사로 표시된 문자열에 대해서는 매칭이 없어야 한다.

힌트: 소괄호와 파이프 기호 `|`를 함께 사용하면 여러 개의 문자열중에 하나와 매칭되는 문자열을 캡처한다.

| 매칭 | 문자열 |
| :---: | :--- |
| 성사 | `tomato: 1` |
| 성사 | `potatos: 2` |

답:

```python
match = re.search(r'(tom|pot)atos?: (\d)', 'tomato: 1')
match
```

```python
match.group(0)
```

```python
match.group(1)
```

```python
match.group(2)
```

```python
match = re.search(r'(tom|pot)atos?: (\d)', 'potatos: 2')
match
```

```python
match.group(0)
```

```python
match.group(1)
```

```python
match.group(2)
```

여는 소괄호 바로 다음에 `?:`를 추가하면 캡처는 지정되지 않는다.

```python
match = re.search(r'(?:tom|pot)atos?: (\d)', 'potatos: 2')
match
```

```python
match.group(0)
```

```python
match.group(1)
```

**문제 7**

전화번호 끝 네자리를 별표로 처리하는 코드를 작성하라.
단 정규표현식을 컴파일한 패턴과 패턴의 `sub()` 메서드만 이용한다.

```python
phone_numbers = """
010-8009-9118
031-700-1919
02-769-9877
"""
```

답:

```python
pattern = re.compile(r"(\d{2,3})-(\d{3,4})-(\d{4})")
print(pattern.sub(r"\g<1>-\g<2>-****", phone_numbers))
```

```python
re.search(regex, '3.14529')
```

```python
re.search(regex, '-255.34')
```

```python
re.search(regex, '128')
```

```python
re.search(regex, '1.9e10')
```

```python
re.search(regex, '123,340.00')
```

반면에 미성사로 표시된 문자열과는 매칭되지 않는다.

```python
re.search(regex, '720p')
```

선언한 다음에 이를 확인하라.
단, 이메일 주소의 아이디는 캡처되어야 한다.
또한 정규표현식을 컴파일한 패턴을 이용한다.

| 그룹 지정 | 문자열 | 그룹 |
| :---: | :--- | :--- |
| 성사 | `tom@hogwarts.com` | `tom` |
| 성사 | `tom.riddle@hogwarts.com` | `tom.riddle` |
| 성사 | `tom.riddle+regexone@hogwarts.com` | `tom.riddle` |

답:

먼저 정규표현식을 컴파일한다.

```python
pattern = re.compile(r'^([\w\.]+)(\+[\w\.]+)*@[\w\.]+')
```

`regex` 변수가 가리키는 정규표현식이 성사로 표시된 모든 문자열 전체와 매칭되며
이메일 주소의 아이디를 캡처한다.

```python
match = pattern.search('tom@hogwarts.com')
match
```

```python
match.group(1)
```

```python
match = pattern.search('tom.riddle@hogwarts.com')
match
```

```python
match.group(1)
```

```python
match = pattern.search('tom.riddle+regexone@hogwarts.com')
match
```

```python
match.group(1)
```
