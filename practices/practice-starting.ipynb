### 문제 5

오전 6시 52분에 집을 떠나서,
초반에는 km당 8분 15초의 속력으로 1km를 뛰고, 
이어서 km당 7분 12초의 속력으로 3km를 뛴 후, 
초반 속력으로 1km를 마저 달려 집으로 돌아왔을 때의 도착시간을 계산하는 코드를 작성하라.
단 집에 도착하는 시간을 아래와같이 출력되도록 해야 한다.

```
집에 도착 시간: 7 시 30 분 6 초
```

힌트: 시, 분, 초를 동시에 다루기 위해 계산을 초 단위로 통일시킨 후 아래 변수들을 선언하고 최종적으로 도착 시간을 지정된 형식으로 출력한다.

- `start`: 자정 기준 출발 시간까지 지난 시간
- `running_time`: 달린 시간
- `arrival_time`: 자정 기준 도착 시간까지 지난 시간

답:

(1) 초단위 도착시간 계산

시, 분, 초를 별도로 다루면 계산이 복잡해지기에 모든 값을 초 단위로 통일시킨다.
먼저 출발시간인 6시 52분이 밤 0시를 기준으로 몇 초 지난 시간인지
계산해서 `start` 변수에 할당한다.

````python
start = (6 * 60 + 52) * 60
print(start)
````

아래 두 변수는 1 킬로미터를 가볍게 달릴 때와 좀 빠르게 달릴 때 걸리는 시간(초)을 가리킨다.

````python
slow_running = 8 * 60 + 15 # 가볍게 달릴 때 킬로미터 당 걸리는 시간(초)
fast_running = 7 * 60 + 12 # 좀 빠르게 달릴 때 킬로미터 당 걸리는 시간(초)
````

그러면 달리는 시간을 초로 계산하면 2286초로 나온다.

````python
running_time = slow_running + 3 * fast_running + slow_running
running_time
````

도착시간은 출발시간에 달린 시간을 더하면 된다.

````python
arrival = start + running_time
arrival
````

(2) 시, 분, 초로 변환

이제 도착시간을 시, 분, 초로 변환하기 위해 1분은 60초, 1시간은 60분이라는 정보를
이용하여 도착시간에 대해 60으로 나눴을 때의 몫과 나머지 계산을 두 번 연속 적용한다.

- 도착시간의 초 계산: 60으로 나눴을 때의 나머지

````python
arrival_seconds = arrival % 60
arrival_seconds
````

- 도착시간의 분 계산: 60으로 나눴을 때의 몫

````python
arrival_minutes = arrival // 60
arrival_minutes
````

450분은 60분 보다 크기에 시와 분으로 다시 변환해야 한다.
한 시간이 60분이기에 역시 60으로 나눈 몫과 나머지를 계산한다.

- 도착시간의 시 계산: 60으로 나눴을 때의 몫

````python
arrival_hours = arrival_minutes // 60
arrival_hours
````

- 도착시간의 분 계산: 60으로 나눴을 때의 나머지

````python
arrival_minutes = arrival_minutes % 60
arrival_minutes
````

따라서 도착시간은 7시 30분 6초가 된다. 

````python
print("집에 도착 시간:", arrival_hours, "시", arrival_minutes, "분", arrival_seconds, "초")
````

(3) 종합 정리

지금까지 설명한 모든 코드를 정리하면 다음과 같다.

````python
start = (6 * 60 + 52) * 60
slow_running = 8 * 60 + 15 # 가볍게 달릴 때 킬로미터 당 걸리는 시간(초)
fast_running = 7 * 60 + 12 # 좀 빠르게 달릴 때 킬로미터 당 걸리는 시간(초)

running_time = (1 + 1) * slow_running + 3 * fast_running
arrival = start + running_time

arrival_seconds = arrival % 60
arrival_minutes = arrival // 60
arrival_hours = arrival_minutes // 60
arrival_minutes = arrival_minutes % 60

print("집에 도착 시간:", arrival_hours, "시", arrival_minutes, "분", arrival_seconds, "초")
````
