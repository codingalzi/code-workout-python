**문제 1**

아래 코드의 실행 결과를 설명하라.

```python
print("\tHello World\n\\")
```

답:

세 종류의 화이트 스페이스가 사용되었음. 

- `' '`: 공백
- `'\t'`: 탭
- `'\n'`: 줄바꿈

문자열이 먼저 탭(스페이스 4칸)으로 시작하고 이후에 Hello World를 출력한 다음 줄바꿈을 진행한다.
줄이 바뀐 후 백슬래시(원화) 기호가 이스케이프된 상태로 출력된다.

**문제 2**

아래 코드의 실행 결과를 설명하라.

```python
len(' \t\n' * 10)
```

답:

사용된 세 개의 모든 화이트스페이스 문자의 길이는 1이다.
문자열 `' \t\n'`의 길이는 3인데 이 문자열을 10번 복제해서 이어붙인 결과의 문자열은 따라서 30이다.

**문제 3**

아래 세 코드의 실행 결과의 차이점을 설명하라.

- 코드 1

```python
print(r"C:\Users\user\Downloads")
```

- 코드 2

```python
print(r"C:\Users\user\Downloads\")
```

- 코드 3

```python
print(r"C:\Users\user\Downloads\\")
```

답:

- `r"C:\Users\user\Downloads")`: 디렉토리의 경로를 보여주는 날 문자열. 정상적으로 있는 그대로 출력됨.

- `r"C:\Users\user\Downloads\")`: 문자열 오른쪽 맨 끝에 위치한 이스케이프가 문자열의 끝을 알리는 큰따옴표에 영향을 주기 때문에 문자열의 끝이 결정되지 못해 오류가 발생한다.

- `r"C:\Users\user\Downloads\\")`: 이전 문자열과는 달리 오른쪽 맨 끝에 위치한 두 개의 이스케이프 문자가 사용됨. 따라서 마지막의 닫는 인용부호에 영향을 주지 않아 있는 그대로 화면에 출력됨.

**문제 4**

아래 부동소수점을 소수점 이하 셋째 자리까지만 출력되도록 f-문자열을 활용하라.

```python
num = 23.36778
num1 = -84.99888
```

답:

아래 코드는 소수점 이하 넷째 자리에서 반올림한 값을 출력한다.

```python
print(f"{num:.3f}")
print(f"{num1:.3f}")
```

**문제 5**

아래 부동소수점의 소수점을 모두 버리로 정수 부문만 출력되도록 f-문자열을 활용하라.
또한 정수는 총 10칸을 차지하도록 한다.
단, `int()` 함수는 사용하지 않는다.

```python
num2 = 23.36778
num3 = 8223.62874636778
```

답:

아래 코드는 소수점 이하 첫째 자리에서 반올림한 값을 출력하며,
총 10칸을 차지하도록 하며 오른쪽으로 정렬시킨다.

```python
print(f"{num2:10.0f}")
print(f"{num3:10.0f}")
```

**문제 6**

아래 두 개의 부동소수점을 총 15 칸을 차지하도록 f-문자열을 활용하라.
또한 소수점 이하 다섯째 자리까지만 출력되록 한다.
단 모든 정수는 오른쪽으로 정렬되어야 하며 앞쪽 빈칸은 `+` 기호로 채운다.

```python
num4 = 887223.8348161
num5 = 6833.127230009267
```

답:

아래 코드는 소수점 이하 여섯째 자리에서 반올림한 값을 출력하며,
총 15칸을 차지하도록 하며 오른쪽으로 정렬시킨다.
또한 왼쪽의 빈 칸은 `+` 기호로 채운다.

```python
print(f"{num3:+>15.5f}")
print(f"{num4:+>15.5f}")
```
