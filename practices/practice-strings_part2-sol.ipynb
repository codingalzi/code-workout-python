**문제 2**

아래 `current_time` 변수는 시각을 서양식으로 표현된 문자열을 가리킨다.

````python
current_time = 'Fri Mar 21, 2025'
````

`current_time` 변수가 가리키는 문자열을 한국식, 즉 다음과 같이 출력하는 코드를 작성한다.

```python
날짜: 2025년 3월 21일 금요일
```

단, 아래 코드에서 선언된 세 개의 변수와 `split()` 메서드, f-문자열을 반드시 활용한다.

````python
weekdays_en = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
weekdays_kr = ['일', '월', '화', '수', '목', '금', '토']
months_en = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
````

답:

먼저 문자열을 쪼개 각각의 정보를 얻는다.

````python
ctime_split = current_time.split()
ctime_split
````

그런 다음 리스트 인덱싱을 이용하여 한국식으로 출력해본다.

````python
weekday = ctime_split[0]
month = ctime_split[1]
day = ctime_split[2]
year = ctime_split[-1]

print(f"날짜: {year}년 {month}월 {day}일 {weekday}")
````

실행 결과에 두 가지 문제가 있다.

- 날짜에 쉼표가 포함됨.
- 영어 표현이 그대로 사용됨.

먼저 날짜에 포함된 쉼표를 삭제하자.

````python
day = day.strip(',')
day
````

좀 더 좋아졌다.

````python
print(f"{year}년 {month}월 {day}일 {weekday}")
````

이제 요일은 한글로, 월은 숫자로 변환한다.

````python
weekday_index = weekdays_en.index(weekday)
weekday_kr = weekdays_kr[weekday_index]
weekday_kr
````

영어식 월 표현을 숫자로 변환하는 것은 보다 간단하다.

````python
month_kr = months_en.index(month) + 1
month_kr
````

위 내용을 이용하여 원하는 결과를 출력할 수 있다.

````python
weekday_index = weekdays_en.index(weekday)
weekday_kr = weekdays_kr[weekday_index]

month_kr = months_en.index(month) + 1

print(f"날짜: {year}년 {month_kr}월 {day}일 {weekday_kr}요일")
````

**문제 3**

변수와 함수의 이름을 지정할 때 **낙타**<font size = "2">camel</font> 표기법 또는
**팟홀**<font size="2">pothole</font> 표기법을 사용한다.

* 낙타 표기법: 소문자로 시작하고, 이어지는 단어의 시작은 대문자로 작성하는 표기법.
* 팟홀 표기법: 모두 소문자를 사용하고, 단어 사이에 밑줄기호(`_`)를 사용하는 표기법.

활용 예제는 다음과 같다.

- 낙타 표기법 예제: `userName`, `printMessage`, `countA` 등.
- 팟홀 표기법 예제: `user_name`, `print_message`, `count_a` 등.   

(1) 아래에 팟홀 표기법을 사용하는 문자열을 지정한다.

````python
counting = 'one_two_three'
````

위 문자열을 활용하여 낙타 표기법을 사용하는 문자열을 생성하라.
단 모든 단어의 첫글짜는 다음과 같이 대문자로 시작해야 한다.

```
OneTwoThree
```

힌트: `split()`, `join()`, `title()` 메서드 활용

답:

밑줄 기준으로 `split()` 메서드를 적용하면 다음과 같이 작동한다.

````python
counting_space = ' '.join(counting.split('_'))
counting_space
````

이제 `split()`과 `join()` 메서드를 적절히 활용하면 원하는 결과를 얻는다.

````python
''.join(counting_space.title().split())
````

(2) 낙타 표기법으로 작성된 문자열이 다음과 같다.

````python
ott_camel = 'OneTwoThree'
````

위 문자열을 팟홀 표기법으로 변환하는 코드를 작성하라.

힌트: `for`문자열과 `isupper()`메서드 활용

답:

아래 함수는 주어진 문자열에 포함된 알파벳을 대상으로
소문자인 경우는 그대로,
대문자인 경우엔 소문자로 변환한 다음 밑줄과 함께 연결하는
방식으로 새로운 문자열을 생성한다.

해당 문자가 대문자인지 여부는 문자열 `isupper()` 메서드로 판단하며,
첫 문자는 소문자로 변환한 상태로 시작한다.

````python
# 첫 문자는 소문자로 지정
ott_pothole = ott_camel[0].lower()

# 이후에 나오는 대문자는 밑줄과 함께 소문자로 연결
for char in ott_camel[1:]:
    if char.isupper():
        ott_pothole += "_" + char.lower()
    else:
        ott_pothole += char

print(ott_pothole)
````

**문제 4**

**어구전철**<font size='2'>anagram</font>은 주어진 단어에 사용된 철자의 순서를 변경해서 생성되는 단어를 가리킨다.
예를 들어 "python"과 "thpyon" 서로 어구전철 관계다.

`'happ y'`와 `' ypp ah'` 두 문자열에 대해 공백을 무시하면서 어구전철 여부를 판단하는 코드를 작성하라.

답:

`sorted()` 함수를 적용하기 전에 `replace()` 메서드를 이용하여 문자열에서 공백을 모두 없애면 된다.

````python
happy1 = 'ha pp y'.replace(' ', '')
happy1
````

````python
happy2 = 'ypp ah'.replace(' ', '')
happy2
````

어구전철 관계의 두 단어는 사용하는 문자의 종류와 빈도가 동일하다.
따라서 두 문자열에 포함된 문자를 알파벳 순으로 정렬했을 때 동일한 값이 된다.

문자열에 포함된 문자를 정렬하기 이행 여기서는 
`sorted()` 함수를 이횽한다.
함수의 반환값으로 문자열에 포함된 알파벳을 크기 순으로 정렬한 리스트가 지정된다.

````python
sorted(happy1) == sorted(happy2)
````

**문제 5**

'기러기', '토마토', '우영우', '인싸 의사의 싸인', 'bob', 'eye', 'pop'와 같이
앞으로 읽어도, 뒤로 읽어도 동일한 문자열을 **회문**<font size='2'>palindrome</font>이라 부른다.

아래 문자열에서 공백 뿐만 아니라 알파벳 대소문자 여부도 무시할 때 회문이 됨을 보이는 코드를 작성하라.

힌트: 문자열 `lower()` 메서드,  문자열 슬라이싱 활용

````python
word = 'A man a plan a canal Panama'
````

답:

이전 문제와 동일한 과정을 밟는다. 
대신 마지막에 `lower()` 메서드를 이용하여 모두 소문자로 변환된 문자열을 이용하여 회문 여부를 판단한다.
회문 여부는 문자열을 바로 읽었을 때와 역순으로 읽었을 때 결과가 동일한지 여부로 판단될 수 있기 때문이다.

````python
word_list = word.split()
word_wo_space = ''.join(word_list)
word_wo_space = word_wo_space.lower()

word_wo_space == word_wo_space[::-1]
````

**문제 6**

영어 소문자 알파벳은 총 26개이다.
`a`, `b`, `c`, ..., `x`, `y`, `z` 각각의 알파벳을 13칸씩 오른쪽으로 이동시키면 다음과 같은 대응 관계가 성립니다.


```
a => n
b => o
c => p
...
x => k
y => l
z => m
```

(1) 변수 `alphabet`가 `'c'`를 가리키도록 하자.

````python
alphabet = 'c'
````

`ord()` 함수와 `chr()` 함수를 이용하여 변수 `alphabet`이 가리키는 값이 대문자 `'P'`가 되도록 하라.

힌트: [문자열 2부](file:///C:/Users/gslee/Documents/GitHub/42H/jupyter-book/_build/html/strings_part2.html)의 예제 참고

답:

영어 알파벳이 26개 이기에 13번째 알파벳인 `m`까지는 `ord()` 값에 13을 더하면 
원하는 알파벳이 지정된다.
실제로 `m`에 대응하는 알파벳은 `z`다.

````python
chr(ord('m') + 13)
````

그런데 대문자 알파벳은 소문자보다 32 작은 값을 갖기에 아래처럼 하면 원하는 결과를 얻는다.

````python
chr(ord('m') + 13 - 32)
````

위 방식을 `alphabet` 변수에 적용하면 대문자 `P`가 나온다.

````python
alphabet = chr(ord(alphabet) + 13 - 32)
print(alphabet)
````
